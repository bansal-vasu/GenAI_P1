# Project Report: Analysis of Dataset and Prediction using Neural Networks

**1. Objectives:**

The primary objective of this project is to develop a predictive model using neural networks to forecast prices of cars based on their various features from a dataset. The goals included:

Understanding the data and performing necessary preprocessing. Building a neural network model, evaluating the model's performance and providing insights into the results.

Understand the distribution of key features like price_in_euro, mileage_in_km, and power_kw.

Analyze the impact of transmission_type and fuel_type on mileage_in_km.

Investigate correlations between different features, such as power_kw, mileage_in_km, and fuel_type.

**2. Data Overview and Preprocessing**

The dataset, gcarf_data.csv, contains various features related to cars, including:
Categorical Features: brand, model, color,  transmission_type, fuel_type, fuel_consumption_l_100km, fuel_consumption_g_km, offer_description	

Numerical Features: registration_date, year, price_in_euro, power_kw, power_ps, mileage_in_km

Data Preprocessing Steps:

Missing Values: The data was checked for any missing values, and rows with missing values were removed to ensure data integrity.

Encoding: Categorical variables were encoded using LabelEncoder to convert them into numerical formats suitable for model training.

**3. Model Development**

The model development process involved creating a simple neural network with the following architecture:

Input Layer: Corresponding to the number of features.

Hidden Layers: Two hidden layers with 64 and 32 neurons, respectively, using ReLU activation functions.

Output Layer: A single neuron to predict the sales, as the task is a regression problem.

Dropout: Dropout layers were added to the model to prevent overfitting by randomly dropping neurons during the training process.

The model was compiled using the Adam optimizer and Mean Squared Error (MSE) as the loss function.

**4.Model Training and Evaluation**

The model was trained on 80% of the dataset, with the remaining 20% used for validation.

The training process was tracked, and the following metrics were observed:

Training Loss: The reduction in training loss over epochs indicated that the model was learning effectively.

Validation Loss: The model’s performance on unseen data was monitored using validation loss. 

Results:

The training and validation losses were plotted to analyze the model’s performance. The loss values indicated how well the model was generalizing to new data.

**5. Managerial Insights**

Based on the analysis and model performance, the following insights can be drawn:

1. Relationship Between Price and Power

price_in_euro has a strong positive correlation with both power_kw (0.52) and power_ps (0.52). This suggests that higher-priced cars tend to have higher power output, which may be a key factor for consumers when selecting more expensive models.

The covariance values for price_in_euro with power_kw and power_ps are also high, further confirming the strong relationship.

2. Fuel Consumption and Vehicle Mileage

There is a weak positive correlation between fuel_consumption_l_100km and mileage_in_km (0.008) and a slightly stronger positive correlation with fuel_consumption_g_km (0.095). This indicates that fuel consumption is not strongly related to the mileage of the vehicle, suggesting that higher mileage does not necessarily lead to higher fuel consumption.

The covariance between fuel_consumption_l_100km and mileage_in_km is relatively low, which supports the observation from the correlation matrix.

3. Year of Manufacture

year shows a moderate negative correlation with mileage_in_km (-0.72). This implies that older vehicles generally have higher mileage, which could reflect their age and usage history.

The negative covariance between year and mileage_in_km also confirms this trend.

4. Transmission and Fuel Type

transmission_type has a moderate positive correlation with fuel_type (0.22), indicating that different transmission types might be associated with specific fuel types.

The covariance values between transmission_type and fuel_type are positive but not very high, which aligns with the observed correlation.

5. Fuel Consumption Metrics

fuel_consumption_l_100km and fuel_consumption_g_km have a strong positive correlation (0.68). This suggests that the fuel consumption in liters per 100 km and grams per km are closely related, reflecting consistency in fuel consumption measurements.

The high covariance between fuel_consumption_l_100km and fuel_consumption_g_km supports the strong correlation, showing that as fuel consumption in one metric increases, it tends to increase in the other metric as well.

6. Impact of Car Brand and Model

The brand and model have a very high positive correlation (0.98), which is expected as the brand and model are often strongly linked. However, this does not impact other variables significantly.

The covariance between brand and model is also very high, reinforcing the strong relationship between these two attributes.

**6. Neural Network Insights**

General Trend:

The validation loss is generally decreasing over the epochs, indicating that the model is learning and improving its performance on the validation data.

Loss Fluctuations:

There is some fluctuation in the training loss between epochs, with some epochs showing higher loss compared to others. This could be a sign of instability or potential overfitting.

Convergence:

The validation loss decreases consistently, which is a good sign that the model is converging. However, the training loss does not show a consistent trend of decrease, which might suggest that further fine-tuning of the learning rate or other hyperparameters could be beneficial.

Potential Overfitting:

If the training loss continues to decrease but the validation loss starts to increase, it could indicate overfitting. However, since the validation loss is also decreasing, this is not currently a major concern.

Overall, the training process seems to be progressing well with improvements in validation loss. Fine-tuning the model and hyperparameters could help in achieving better performance.

## Neural Network

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
print(df.dtypes)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
print(df.dtypes)
```

```python
# Define features (X) and target (y)
X = df.drop('price_in_euro', axis=1)  # Use other columns as features
y = df['price_in_euro']  # Predicting price_in_euro

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python

# Create the neural network
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))  # Output layer for regression (predicting a single continuous value)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python

# Evaluate the model on the test set
loss = model.evaluate(X_test, y_test)

# Make predictions
predictions = model.predict(X_test)

# Output the loss and first few predictions
print("Loss:", loss)
print("First 5 Predictions:", predictions[:5])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```
